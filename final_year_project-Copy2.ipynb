```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from math import ceil

class DeepEfficientModelWithResidual(nn.Module):
    def __init__(self, measurement_rate=0.25):
        super(DeepEfficientModelWithResidual, self).__init__()

        self.measurement_rate = measurement_rate
        self.fc1 = nn.Linear(ceil(self.measurement_rate * 1089), 500)
        self.fc2 = nn.Linear(500, 1089)  

        self.conv1 = nn.Conv2d(1, 32, 3, 1, padding=1)
        self.conv2 = nn.Conv2d(32, 64, 3, 1, padding=1)
        self.conv3 = nn.Conv2d(64, 64, 3, 1, padding=1)
        self.conv4 = nn.Conv2d(64, 64, 3, 1, padding=1)
        self.conv5 = nn.Conv2d(64, 64, 3, 1, padding=1)
        self.conv6 = nn.Conv2d(64, 64, 3, 1, padding=1)
        self.conv7 = nn.Conv2d(64, 64, 3, 1, padding=1)
        self.conv8 = nn.Conv2d(64, 32, 3, 1, padding=1)
        self.conv9 = nn.Conv2d(32, 16, 3, 1, padding=1)
        self.conv10 = nn.Conv2d(16, 8, 3, 1, padding=1)
        self.conv11 = nn.Conv2d(8, 1, 3, 1, padding=1)

        self.dropout = nn.Dropout2d(p=0.2)
        self.leaky_relu = nn.LeakyReLU(0.2)

    def forward(self, x):
        # Sparse representation learning
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = x.view(-1, 33, 33)
        x = x.unsqueeze(1)
        
        # Convolutional layers for feature extraction with residual connections
        residual = x
        x = self.leaky_relu(self.conv1(x))
        x = self.leaky_relu(self.conv2(x) + residual)
        residual = x
        x = self.leaky_relu(self.conv3(x))
        x = self.leaky_relu(self.conv4(x) + residual)
        residual = x
        x = self.leaky_relu(self.conv5(x))
        x = self.leaky_relu(self.conv6(x) + residual)
        residual = x
        x = self.leaky_relu(self.conv7(x))
        x = self.leaky_relu(self.conv8(x) + residual)
        x = self.leaky_relu(self.conv9(x))
        x = self.leaky_relu(self.conv10(x))
        
        # Reconstruction layer
        x = self.conv11(x)

        return x
```

```python
# inp (images) - is the compressed file
# label (labels) - is the original file
# The compressed image size, in the context of compressive sensing,
# usually refers to a reduced set of measurements or features that represent the original image.
# The goal of compressive sensing is to represent the image using fewer measurements than the number
# of pixels in the original image.
```

```python
import time
total_start_time = time.time()
def train(model, criterion, optimizer, train_dl, num_epochs=10):
    for epoch in range(num_epochs):
        start_time = time.time()
        train_loss = []
        model.train()
        for inp, lbl in train_dl:
            inp = inp.cuda()
            lbl = lbl.cuda()
            inp = inp.float()
            lbl = lbl.float()

            optimizer.zero_grad()
            out = model(inp)
            out = out.cuda()
            out = out.view(lbl.size())
            loss = criterion(out, lbl)
            loss.backward()
            optimizer.step()

            train_loss.append(loss.item())

        epoch_time = time.time() - start_time
        total_time = time.time() - total_start_time  # Total time taken for all epochs so far

        print(f'Epoch: {epoch+1}, Training Loss: {np.mean(train_loss)}, Time taken: {epoch_time:.2f} seconds, Total time taken: {total_time:.2f} seconds')


train(model, criterion, optimizer, train_dl, 10)
```

```python
image_blocks = []
test_img = cv2.cvtColor(test_img, cv2.COLOR_BGR2YCR_CB)
test_img = cv2.normalize(test_img, None, alpha=0, beta=1, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_32F)
test_img = test_img[:,:,0]
h,w = test_img.shape
stride = 14
filter_size = 33
h_n = ((h - filter_size) // stride) + 1
w_n = ((w - filter_size) // stride) + 1

for i in range(h_n):
  for j in range(w_n):
    blocks = test_img[i*stride:(i*stride)+filter_size, j*stride:(j*stride)+filter_size]
    image_blocks.append(blocks)

image_blocks = np.array(image_blocks)
```

```python
import numpy as np

def psnr(img1, img2):
    mse = np.mean((img1 - img2) ** 2)
    if mse == 0:
        return float('inf')
    PIXEL_MAX = 255.0  # Assuming pixel values are in the range [0, 255]
    return 20 * np.log10(PIXEL_MAX / np.sqrt(mse))
```
