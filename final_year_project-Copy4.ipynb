```python
def normalize(v):
    return v / np.sqrt(v.dot(v))

def generate_phi(x, y):
    np.random.seed(333)
    phi = np.random.normal(size=(x, y))
    n = len(phi)
    
    # perform Gramm-Schmidt orthonormalization
    
    phi[0, :] = normalize(phi[0, :])
    
    for i in range(1, n):
        Ai = phi[i, :]
        for j in range(0, i):
            Aj = phi[j, :]
            t = Ai.dot(Aj)
            Ai = Ai - t * Aj
        phi[i, :] = normalize(Ai)
        
    return phi

mat = generate_phi(272, 1089)
mat = torch.from_numpy(mat)
mat
```

```python
import torch
import time
import numpy as np
import os
from skimage.metrics import peak_signal_noise_ratio as psnr

# Create a directory to save models
os.makedirs('saved_models', exist_ok=True)

total_start_time = time.time()

def train(model, criterion, optimizer, train_dl, num_epochs=10, scheduler=None, early_stopping_patience=10):
    best_train_loss = float('inf')
    patience_counter = 0

    # Move model to CUDA device once
    model = model.cuda()

    for epoch in range(num_epochs):
        start_time = time.time()
        train_loss = []
        psnr_values = []
        model.train()

        for inp, lbl in train_dl:
            inp = inp.cuda().float()
            lbl = lbl.cuda().float()

            optimizer.zero_grad()
            out = model(inp)
            out = out.view(lbl.size())
            loss = criterion(out, lbl)
            loss.backward()
            optimizer.step()

            train_loss.append(loss.item())

            # Calculate PSNR
            out_cpu = out.detach().cpu().numpy()
            lbl_cpu = lbl.detach().cpu().numpy()
            psnr_values.append(psnr(lbl_cpu, out_cpu))

        epoch_time = time.time() - start_time
        total_time = time.time() - total_start_time  # Total time taken for all epochs so far

        mean_train_loss = np.mean(train_loss)
        mean_psnr = np.mean(psnr_values)

        print(f'Epoch: {epoch+1}/{num_epochs}, Training Loss: {mean_train_loss:.6f}, PSNR: {mean_psnr:.2f}, Time taken: {epoch_time:.2f} seconds, Total time taken: {total_time:.2f} seconds')

        # Save the best model
        if mean_train_loss < best_train_loss:
            best_train_loss = mean_train_loss
            patience_counter = 0
            model_path = os.path.join('saved_models', f'best_model_epoch_{epoch+1}.pth')
            torch.save(model.state_dict(), model_path)
            print(f"Saved best model at epoch {epoch+1} with training loss: {mean_train_loss:.6f}")
        else:
            patience_counter += 1
            if patience_counter >= early_stopping_patience:
                print("Early stopping triggered")
                break

        # Learning rate scheduler
        if scheduler:
            scheduler.step()

    total_duration = time.time() - total_start_time
    print(f"Training completed in: {total_duration:.2f} seconds")
```

```python
# Pass the scheduler to train function
train(model, criterion, optimizer, train_dl, num_epochs=200, scheduler=scheduler, early_stopping_patience=10)
```

```python
import numpy as np
from PIL import Image

def crop_image(image, num_pixels_right, num_pixels_bottom):
    """
    Crops the specified number of pixels from the right and bottom sides of the image.

    Args:
    image (PIL.Image or np.ndarray): Input image to be cropped.
    num_pixels_right (int): Number of pixels to crop from the right side.
    num_pixels_bottom (int): Number of pixels to crop from the bottom side.

    Returns:
    PIL.Image or np.ndarray: Cropped image.
    """
    # Convert the image to a numpy array if it's not already
    if isinstance(image, Image.Image):
        image = np.array(image)

    # Get the dimensions of the image
    height, width = image.shape[:2]

    # Calculate the new dimensions
    new_width = width - num_pixels_right
    new_height = height - num_pixels_bottom

    if new_width <= 0 or new_height <= 0:
        raise ValueError("Number of pixels to crop exceeds image dimensions")

    # Crop the image
    cropped_image = image[:new_height, :new_width]

    # Convert back to PIL image if the input was a PIL image
    if isinstance(image, Image.Image):
        cropped_image = Image.fromarray(cropped_image)

    return cropped_image

```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def rgb_to_ycbcr(image):
    """
    Convert an image from RGB to YCbCr color space.
    
    Parameters:
    - image: numpy array, input RGB image
    
    Returns:
    - ycbcr_image: numpy array, output YCbCr image
    """
    ycbcr_image = cv2.cvtColor(image, cv2.COLOR_RGB2YCrCb)
    return ycbcr_image

def ycbcr_to_rgb(image):
    """
    Convert an image from YCbCr to RGB color space.
    
    Parameters:
    - image: numpy array, input YCbCr image
    
    Returns:
    - rgb_image: numpy array, output RGB image
    """
    rgb_image = cv2.cvtColor(image, cv2.COLOR_YCrCb2RGB)
    return rgb_image

# Example usage:
# Load an RGB image
rgb_image = cv2.imread('robot.jpg', cv2.IMREAD_COLOR)

# Convert RGB to YCbCr
ycbcr_image = rgb_to_ycbcr(rgb_image)

# Convert YCbCr back to RGB
rgb_image_restored = ycbcr_to_rgb(ycbcr_image)

# Display the original and restored images using matplotlib
plt.figure(figsize=(10, 5))

# Original RGB image
plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(rgb_image, cv2.COLOR_BGR2RGB))
plt.title('Original RGB Image')
plt.axis('off')

# YCbCr image
plt.subplot(1, 3, 2)
plt.imshow(ycbcr_image)
plt.title('YCbCr Image')
plt.axis('off')

# Restored RGB image
plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(rgb_image_restored, cv2.COLOR_BGR2RGB))
plt.title('Restored RGB Image')
plt.axis('off')

plt.show()
```

```python
import cv2
import numpy as np
from skimage.restoration import denoise_nl_means, denoise_tv_chambolle
import pywt
import tensorflow as tf
import torch

# Non-local Means (NLM) Denoising
def nlm_denoise(image):
    denoised_image = denoise_nl_means(image, h=0.8, fast_mode=True, patch_size=5, patch_distance=3)
    return denoised_image

# Wavelet-based Denoising
def wavelet_denoise(image):
    coeffs = pywt.dwt2(image, 'haar')
    coeffs = tuple(denoise_tv_chambolle(coeff, weight=0.1) for coeff in coeffs)
    denoised_image = pywt.idwt2(coeffs, 'haar')
    return denoised_image

# Total Variation (TV) Denoising
def tv_denoise(image):
    denoised_image = denoise_tv_chambolle(image, weight=0.1)
    return denoised_image

# Gaussian Mixture Model (GMM) Denoising (OpenCV implementation)
def gmm_denoise(image):
    denoised_image = cv2.fastNlMeansDenoisingColored(image, None, 10, 10, 7, 21)
    return denoised_image

# Deep Learning-based Denoising (TensorFlow implementation)
def deep_learning_denoise(image):
    model = tf.keras.models.load_model('denoising_model.h5')  # Load pre-trained model
    denoised_image = model.predict(np.expand_dims(image, axis=0))
    return denoised_image[0]

# Deep Learning-based Denoising (PyTorch implementation)
def deep_learning_denoise_pytorch(image):
    model = torch.load('denoising_model.pth')  # Load pre-trained model
    model.eval()
    image_tensor = torch.from_numpy(np.moveaxis(image, -1, 0)).float().unsqueeze(0)
    with torch.no_grad():
        denoised_image = model(image_tensor).squeeze().numpy()
    return np.moveaxis(denoised_image, 0, -1)
```
